```python
import pandas as pd
```

```python
df = pd.read_csv('suicide.csv')
```

## Questions:

### 1. Display Country, Year, Sex, Age, Suides/100k pop columns

```python
df[['country','year','sex','age','suicides_no','suicides/100k pop']]
```

### 2. Display the Philippines suicide record. Display Year, Sex, Age, suicides_n, and Suides/100k pop columns.

```python
df[df['country'] == 'Philippines'][['year','sex','age','suicides_no','suicides/100k pop']]
```

### 3. Display the Philippines suicide record in year 2011. Display Year, Sex, Age, suicides_n, and Suides/100k pop columns.

```python
df[(df['country'] == 'Philippines') & (df['year'] == 2011)][['sex','age','suicides_no','suicides/100k pop']]
```

### 4. Display the country with the highest suicide per 100k population in 2005.Display Year, Sex, Age, and Suides/100k pop columns


```python
max_dpop = df[df['year'] == 2005]['suicides/100k pop'].max()
```

```python
df[df['suicides/100k pop'] == max_dpop][['country','year','sex','age','suicides_no','suicides/100k pop']]
```

### 5. Display the total suicide cases (suicides_n) and suides/100k pop per year. Sort in descending order based on suicides_n.


```python
df.groupby(['year'])[['suicides_no','suicides/100k pop']].sum().sort_values(by='suicides_no',ascending=False)
```

### 6. Display the total suicide cases (suicides_n) and suides/100k pop per gender. Sort in descending order based on suicides_n.


```python
df.groupby(['sex'])[['suicides_no','suicides/100k pop']].sum().sort_values(by='suicides_no',ascending=False)
```

### 7. Display the total suicide cases (suicides_n) and suides/100k pop per age. Sort in descending order based on suicides_n.


```python
df.groupby(['age'])[['suicides_no','suicides/100k pop']].sum().sort_values(by='suicides_no',ascending=False)
```

### 8. Display the total suicide cases (suicides_n) per country and per year. Sort in descending order based on suicides_n.


```python
df.groupby(['country','year'])[['suicides_no']].sum().sort_values(by='suicides_no',ascending=False)
```

### 9. Display the total suicide cases in the Philippines based on year, gender, and age.


```python
df[df['country'] == 'Philippines'].groupby(['year','sex','age']).sum()
```

### 10. Display the year and number of suicide cases that the Philippines has the highest number of combined suicide cases.

```python
df[df['country']=='Philippines'].groupby(['year']).sum().reset_index().sort_values(by='suicides_no',ascending=False).iloc[0][['year','suicides_no']]
```
