# Online Business Sales Analysis (2017-2019)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('business.retailsales2.csv')
```

```python
df.head()
```

```python
df['Month'] = pd.to_datetime(df['Month'],format='%B').dt.month

df
```

### What is the total number of orders placed in each year?


```python
df.groupby(['Year'])['Total Orders'].sum()
```

### What is the average value of discounts given in each month?


```python
df.groupby(['Month'])['Discounts'].mean()
```

### What is the monthly net sales revenue trend over the last two years?


```python
monthly = df.loc[df['Year'] > 2017].groupby(['Month'])['Net Sales'].sum().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(monthly['Month'], monthly['Net Sales'])
ax.set_xlabel('Month')
ax.set_ylabel('Net Sales')
ax.set_title('Monthly Net Sales Revenue Trend')
plt.show()
```
