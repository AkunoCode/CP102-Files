```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv('restaurant_orders.csv')

sales_data.head()
```

### What is the total revenue generated by the restaurant in the last year?

```python
# Create 'Sales Amount' Column for each order
sales_data["Sales Amount"] = sales_data['Quantity'] * sales_data['Price'] - (sales_data['Price'] * sales_data['Discount'])

sales_data.head()
```

```python
revenue = sales_data["Sales Amount"].sum()

print(f"The Restaurant's Revenue is P{revenue:.2f}")
```
